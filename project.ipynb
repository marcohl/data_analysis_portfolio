```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd

df = pd.read_csv('datasets/lego_sets.csv')
parent_theme = pd.read_csv('datasets/parent_themes.csv')
```

```python
df.info()
```

```python
merge = pd.merge(df, parent_theme, left_on = 'parent_theme', right_on = 'name')
```

```python
merge.head()
```

```python
merge.info()
```

```python
merge.drop('name_y', axis=1, inplace= True)
```

```python
merge.head()
```

```python
star_wars = merge[merge['parent_theme'] == 'Star Wars']
```

```python
licensed_prod = merge[merge['is_licensed'] == True]
licensed_prod = licensed_prod.dropna(subset=['set_num'])
```

```python
licensed_prod.info()
```

```python
the_force = int((len(star_wars)/len(licensed_prod)*100))
print(the_force)
```

```python
licensed_prod.head()
```

```python
grouped = licensed_prod.groupby(['year','parent_theme']).sum()
```

```python
grouped.head()
```

```python
most_popular_year = grouped.loc[grouped.groupby('year')['is_licensed'].idxmax()]
most_popular_year
```
